```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_train_flatten=x_train.reshape(len(x_train),28*28)
x_train_flatten.shape
```

```python
x_train_flatten[0]
```

```python
x_test_flatten=x_test.reshape(len(x_test),28*28)
x_test_flatten.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaling=MinMaxScaler()
x_train_flat_result=scaling.fit_transform(x_train_flatten)
x_train_flat_result[0]
```

```python
x_test_flat_result=scaling.fit_transform(x_test_flatten)
```

```python
x_test_flat_result[0]
```

```python
model=keras.Sequential([
                        keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
                       
                       ])
model.compile(
                optimizer='adam',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy']
             )
model.fit(x_train_flat_result,y_train,epochs=5)
```

```python
 model.evaluate(x_test_flat_result,y_test)
```

```python
plt.matshow(x_test[0])
```

```python
y_predicted=model.predict(x_test_flat_result)
y_predicted[0]
```

```python
np.argmax(y_predicted[0])
```

```python
y_test_labels=[np.argmax(i) for i in y_predicted] 
```

```python
y_test_labels
```

```python
cm=tf.math.confusion_matrix(labels=y_test, predictions=y_test_labels)
```

```python
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True,fmt='d')
```

```python
train_result=[]
scaler=MinMaxScaler()
for i in range(0,len(x_train)):
    scaler.fit(x_train[i])
    train_result.append(scaler.transform(x_train[i]))

```

```python
train_result=np.asarray(train_result)
train_result.shape
```

```python
model=keras.Sequential([
                        keras.layers.Flatten(input_shape=(28,28)),
                        keras.layers.Dense(100,activation='relu'),
                        keras.layers.Dense(10,activation='sigmoid')
                       
                       ])
model.compile(
                optimizer='adam',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy']
             )
model.fit(train_result,y_train,epochs=5)
```
